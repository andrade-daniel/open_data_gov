## loading packages

```python
import requests
import json
import pandas as pd
from IPython.display import display
import warnings

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", None)
```

# setting API key

```python
API_KEY = "INSERT HERE YOUR API KEY"
```

```python
session = requests.Session()
session.headers.update({"X-API-KEY": API_KEY})
```

```python
response = session.get("https://dados.gov.pt/api/1/datasets/")
```

```python
response.json()
```

## loop through all pages

```python
datasets = response.json().get("data")
next_page = response.json().get("next_page")
while next_page:
    response = session.get(next_page)
    datasets += response.json().get("data")
    next_page = response.json().get("next_page")
    print(next_page)
    if not next_page:
        print("Done!")
        break
print("nr of datasets:", len(datasets))
```

```python
# into dataframe
df = pd.DataFrame(datasets)

print(df.shape)
df.head()
```

```python
# Apagar colunas sem qualquer valor
fields_to_del = ("deleted", "private", "extras", "badges")
for field in fields_to_del:
    if field in df:
        del df[field]

# Separar a coluna "temporal_coverage" em campos "coverage_start" e "coverage_end",
# para termos datas limpas
def separate_coverage(d):
    if not d:
        return (None, None)
    return (d.get("start"), d.get("end"))


# https://stackoverflow.com/a/12356541/122400
if "temporal_coverage" in df:
    df["coverage_start"], df["coverage_end"] = zip(
        *df["temporal_coverage"].map(separate_coverage)
    )
    del df["temporal_coverage"]

# Harmonizar datas que estão expressas como datetime
df["frequency_date"] = df["frequency_date"].map(
    lambda x: x.split("T")[0], na_action="ignore"
)
```

```python
print(df.shape)
print(df.head())
```

## extract organization/entity

```python
df["entity"] = [
    org["name"].strip() if org is not None else "" for org in df["organization"]
]
```
